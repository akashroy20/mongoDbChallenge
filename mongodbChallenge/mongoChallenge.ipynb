```python
import pymongo
db_name = input("Enter name of your database: ")
collection_name = input("Enter name of your collection: ")
client = pymongo.MongoClient(input("Enter your connection string: "))
db = client.test
db2 = client[db_name]
coll1 = db2[collection_name]
```

```python
import pandas as pd
l = []
for i in coll1.find():
    l.append(i)
pd.DataFrame(l)
```

```python
ag = coll1.aggregate(
    [{
        '$group':{'_id':'$Chiral indice m','No of records':{'$sum':1}}
    }])
for i in ag:
    print(i)
```

```python
l = []
for i in coll1.find({'Chiral indice n':2.0 , 'Chiral indice m': 1.0}):
    l.append(i)
pd.DataFrame(l)
```

```python
coll2 = db2['Chiral_indice_n_1_Chiral_indice_m_2']
coll2.insert_many(l)
```

```python
ag1 = coll2.aggregate(
    [{
        '$group':{'_id':'$Chiral indice m','No of records having chiral indice n equals 1 and Chiral indice m equals 2':{'$sum':1}}
    }])
for i in ag1:
    print(i)
```

```python
la = []
ag2 = coll2.aggregate(
    [{
        '$group':{'_id':'$Chiral indice m','Mean of initial atomic coordinate u':{'$avg':'$Initial atomic coordinate u'},"mean of calculated atomic coordinates u'":{'$avg':"$Calculated atomic coordinates u'"}}
    }])
for i in ag2:
    la.append(i)
a = pd.DataFrame(la)
a['difference'] = a['Mean of initial atomic coordinate u']-a["mean of calculated atomic coordinates u'"]
a
```
